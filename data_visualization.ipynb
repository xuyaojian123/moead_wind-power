```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def plot_weekdata(data,title='week_data',label='washingtong_winter_20120101-20120107'):
    fig,ax = plt.subplots(figsize=(10, 5),layout='constrained')
    ax.set_title(title)
    ax.set_xlabel('Time')
    ax.set_ylabel('wind speed at 100m (m/s)')
    ax.plot(data,label=label)
    ax.grid()
    ax.legend()
    plt.show()
```

```python
df1 = pd.read_csv('data/week_data/data/washingtong_winter_20120101-20120107新.csv')
```

```python
df1_data = df1['wind speed at 100m (m/s)']
plot_weekdata(df1_data,title='washingtong_winter_20120101-20120107')
```
